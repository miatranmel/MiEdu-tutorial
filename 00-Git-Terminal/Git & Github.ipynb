### 2. Track thay đổi local với Git

Git là công cụ quản lý version của tập tin, hiệu quả trong quá trình phát triển phần mềm. Git được thiết kế để giúp các lập trình viên làm việc chung với nhau một cách hiệu quả trên một code base duy nhất.

#### Các lệnh Git thường được sử dụng

### Quy trình sử dụng Git

#### Tạo project mới

```
mkdir -p ~/code/YOUR_GITHUB_USERNAME/sample_directory && cd $_  
# Create a new folder
git init # Initialise a git repository
ls -al #  A .git hidden folder has automatically been created
```

#### Tạo file và commit thay đổi

```
touch README.md # Create a README.md file
atom .  # Open README file in Atom editor and write something
git status  # The file is not tracked
git add README.md # Track the file
git status  # The file is now tracked
git commit -m "Create README" # Save changes
git status # All changes are saved
```

#### Chỉnh sửa file và commit thay đổi

```
curl https://giphy.com/gifs/Bunim-MurrayProductions-life-kim-kardashian-e-dhjJRAs2ZbWtrYvSUT > kimk.gif # Save a gif from the internet
atom .  # Edit the README file and add kimk.gif
git status  # The changes are not tracked
git diff README.md  # Check what has changed
git add README.md # Track the changes in README.md
git add kimk.gif # Track the new file kimk.gif
git status # All changes are now tracked
git commit -m "Add gif to README.md" # Save changes
git status # All changes are saved
git log  # Check commits history
```

### 3. Làm việc với branches

Giả sử bạn làm 1 project lớn: code của bạn rất clean, chương trình chạy tốt, bạn rất hài lòng. Bây giờ, bạn đang muốn thêm một feature mới vào code base của bạn và để làm feature này bạn cần phải code lại phần lớn code base trước. Ở đây, nếu không dùng Git thì bạn sẽ có nguy cơ làm rối tung code base trước. 🥶🥶🥶

Để giải quyết vấn đề này, Git cung cấp cho bạn tính năng **branches**.

Ý tưởng là bạn sẽ tạo các nhánh khác nhau từ nhánh chính của dự án. Ở đây, bạn sẽ triển khai các phiên bản mã (feature) khác nhau của mình. Sau khi hài lòng với công việc của mình, bạn có thể hợp nhất nhánh chuyển hướng đó vào nhánh chính!

#### 3.1 Tạo phiên bản mã (branch) mới

Tạp phiên bản mã (test-branch) rất dể. Bạn làm theo nhé

```
# Make sure you are on branch master
git checkout master

# Create a branch called `test-branch` and go to that branch
git checkout -b test-branch
```

Sau khi tạo test-branch, bạn có thể tiến hành chỉnh sữa mã và commit trên nhánh phụ test-branch

```
# Create a new file with some stuff written inside it
echo "some stuff" > new_file.txt

# Check your git status : there should be some unstaged/uncommitted changes
git status

# Stage the new file
git add new_file.txt

# Commit the changes with an explicit commit message
git commit -m "Created new_file.txt containing some stuff"
```

Nếu bạn quay trở lại phiên bản mã *master*, bạn sẽ thấy là không có sự thay đổi gì trên branch *master*

```
# Go to the master branch
git checkout master

# no changes to see here !
```

#### 3.2 Hợp nhất branches

Giả sử bây giờ bạn đã chạy test thành công và chắc chắn với những thay đổi trong nhánh *test-branch* và muốn hợp nhất chúng vào nhánh *master*. Bạn có thể dễ dàng làm như sau:

```
# Go to the master branch if you're not already there
git checkout master

# Merge the `test-branch` into master
git merge test-branch

# Delete the merged `test-branch` if you want to (not mandatory)
git branch -d test-branch
```

#### 3.2 Giải quyết conflicts

Đôi khi, khi làm việc với các phiên bản mã (branches), có thể sẽ xãy ra vấn đề là bạn có thể đã sửa đổi cùng một tệp theo hai cách khác nhau trên hai nhánh khác nhau và như vậy, khi cố gắng hợp nhất hai nhánh của bạn vào nhánh chính, bạn sẽ nhận được thông báo như sau:

Trong trong trường hợp này, Git không biết bạn sẽ muốn giữ lại phiên bản nào, do đó Git sẽ báo lỗi. Để giải quyết, bạn cần chọn 1 trong 2 phiên bản hoặc loại bỏ 2 phiên bản và thay thế bằng bằng 1 phiên bản mới. Sau đó bạn thêm (lệnh add) những thay đổi vào file và hợp nhất.

# Git Flow: Quy trình chuẩn làm việc nhóm cùng Github

Khi đi làm, bạn rất hiếm khi làm việc một mình lắm, thường thì bạn sẽ làm việc chung với một nhóm các data scientist hoặc developer khác trên một dự án. Câu hỏi đặt ra là làm thế nào để chúng ta làm việc trơn tru và không dẫm chân lên nhau đây?

Câu trả lời là bạn sử dụng Git Flow, phương pháp được sử dụng rộng rãi bỡi các lập trình viên

#### 1. Git Flow, wow

Okie, để mình giải thích bí kíp hen

**Git Flow** gồm có 2 nhánh chính là Master và Develop, 3 nhánh Phụ gồm: feature, release và hotfix. Mỗi loại nhánh sẽ có chức năng khác nhau.

- Nhánh *master* hay *production ready* là nhánh mặc định trong Git. Đây là branch chính của tất cả các repository → code ở trên branch này là code đã ở trạng thái Production-ready. Các developers không thể tự ý push code lên branch này, mà buộc phải thông qua Pull request từ cách branches khác. Branch sẽ được tự động build và test qua hệ thống Travis-CI (nếu bạn dùng) → nếu build và test failed thì developers cần có trách nhiệm khắc phục càng sớm càng tốt. Code trên branch master sẽ được release lên các *server production* qua cách đánh Tag tương ứng.
- Nhánh *developer* là nhánh trung tâm cho việc phát triển ứng dụng. Mỗi thay đổi bạn nên tách ra thành một nhánh *feature* tương ứng. Mỗi nhánh *feature* tách ra sau khi xong sẽ được hợp nhất vào nhánh *developer*, xong sẽ xóa đi. Code trên branch develop cũng sẽ được tự động build và test.
- Nhánh *release* dùng cho các phiên bản ứng dụng khác nhau có thời gian tồn tại ngắn, được tách ra từ nhánh develop và để chuẩn bị cho việc release (ví dụ: thay version number). Sau khi hoàn thành các thay đổi, branch sẽ được merge lại vào develop và master để tiến hành release (= cách đánh tag trên master). Branch này sau đó sẽ được remove ngay lập tức. Tên branch release nên được đặt theo tên version, ví dụ: release-v1.0.0-beta1
- Nhánh *feature* là branch dành cho việc phát triển code cho cách tính năng / nhiệm vụ riêng biệt. Branch cần được base từ *develop* và hợp nhất lại vào *develop* (tất cả các pull request để hợp nhất thẳng từ feature vào master sẽ bị reject). Branch này sẽ không được tự động build và test, tuy nhiên developer khi pull request thì hệ thống sẽ build → nếu failed thì developer cần có nghĩa vụ phải fixed trước khi merge vào develop branch. Cách đặt tên branch feature sẽ mặc định là: feature/ten_tinh_nang , feature/ten_nhiem_vu
- Nhánh *hotfix* dùng để fix các tính năng hoặc lỗi. Để fix nhưng không muốn tiến hành một đợt release như workflow ở trên thì ta có thể mở ra những nhánh hotfix (có thể branch bắt đầu từ master). Nhánh này sau khi apply trên master sẽ bắt buộc cần phải merge vào nhánh develop.

#### 2. Versioning tiêu chuẩn

Cách đánh version trong các sản phẩm của DigiMed sẽ theo cách sau:

- *Alpha version*: áp dụng khi đang phát triển nội bộ. Ví dụ: **version 1.0.0-alpha15**
- *Beta version*: áp dụng khi sản phẩm đã được đưa ra production nhưng vẫn được bên khách hàng dùng thử nghiệm. Ví dụ **version 1.0.0-beta15**
- *Release version*: áp dụng khi sản phẩm đã được đưa ra market (nhiều người biết đến, hoặc là sản phẩm thương mại) Version sẽ được đánh đồng nhất trên tất cả các platform (backend, frontend, ios, android). Khi apply hotfixes cho các version, chúng ta sẽ không bump version mà là đánh thêm số, ví dụ đối với **version 1.0.0-alpha15**, khi apply hotfixes sẽ được tags dưới dạng **1.0.0-alpha15.1** , **1.0.0-alpha15.2**

### 3. Thực hành theo Git Flow

Mình giả sử bạn đã khởi tạo thành công nhánh *master* trên Github nhé. 

##### Bước 1: tạo nhánh **develop** theo quy trình Git Flow như sau

```
# Go on the master branch
git checkout master

# Get the latest version of the branch
git pull

# Create a new branch called feature/awesome_feature
git checkout -b developer
```

##### Bước 2: tạo nhánh **feature** mà bạn mong muốn thêm

```
# Go on the developer branch
git checkout developer

# Get the latest version of the branch
git pull

# Create a new branch called feature/awesome_feature
git checkout -b feature/awesome_feature
```

Rồi, giờ bạn code kiếc gì đó cho feature bạn muốn thêm. Bạn có thể an tâm là mình đang làm việc với nhánh *awesome_feature*. Okie, bạn code xong 1 feature mới trong 1 phút 30 giây và hài lòng với kết quả rồi, giờ bạn sẽ đẩy toàn bộ nhánh lên **developer**, ở đó, công việc của bạn sẽ được **SẾP** và các thành viên khác trong nhóm của bạn xem xét. trước khi được hợp nhất vào **developer**. Chú ý, nếu bạn hợp nhất trên nhánh **master**, **SẾP** có thể sẽ mắng bạn tung chảo đấy 😅😅😅

```
# Check unstaged changes in the branch
git status

# Stage the changes
git add my_changed_file.py

# Commit the changes with an explicit commit message
git commit -m ‘The message I want to show to other developers’

# Push the changes to github
git push  #(if pushing for the first time: git push --set-upstream origin feature/awesome_feature)
```

Sau khi đẩy (push) lên nhánh *developer*, bạn sẽ nhận được thông báo mở một **pull request** trên Github như này:

Tuyệt vời, done bước 2, easy job!!!

##### Bước 3: Tạo **pull request** trên Github

Bạn đăng nhập vào Github, tạo *pull request* để thông báo với **SẾP** và mọi người công việc của bạn vừa được đẩy lên *repository*. Một khi *pull request* được mở, mọi người có thể thảo luận và xem xét những thay đổi bạn đã làm hoặc thêm những *commits* khác trước khi hợp nhất vào nhánh developer.
